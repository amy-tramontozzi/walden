```python
from collections import defaultdict
import random

filename = 'walden_clean.txt' 
```

```python
# Question 2
pairs = []

with open(filename, 'r') as file:
    walden = file.read()
    
    sentences = walden.replace('.', ' PERIOD').splitlines()
    
    for sentence in sentences:
        words = sentence.split()
        
        pairs.extend(zip(words, words[1:]))
```

```python
# Question 3, method A
counts = defaultdict(lambda: defaultdict(int))

with open(filename, 'r') as file:
    walden = file.read()
    
    sentences = walden.replace('.', ' PERIOD').splitlines()
    
    for sentence in sentences:
        words = sentence.split()
        # Count pairs
        for i in range(len(words) - 1):
            counts[words[i]][words[i + 1]] += 1
```

```python
# Question 4
probabilities = {}

for word, word_counts in counts.items():
    total = sum(word_counts.values())
    probabilities[word] = {next_word: count / total for next_word, count in word_counts.items()}
```

```python
# Question 5
random.seed(13)

num_sentences = 100
max_sentence_words = 10
starting_word = 'foundation'  

generated_sentences = []

for i in range(num_sentences):
    current_word = starting_word
    sentence = [current_word] # Start each sentence with the same word

    for j in range(max_sentence_words):
        next_words = list(probabilities.get(current_word, {}).keys())
        next_word_probs = list(probabilities.get(current_word, {}).values())
        
        if not next_words:
            break  # No more than 10 words per sentence

        if current_word == "PERIOD":
            break 
        
        # Pick the next word based on first item in the dictionary (highest prob)
        current_word = random.choices(next_words, weights=next_word_probs)[0]
        sentence.append(current_word)

    # Join the sentence and add it to the list
    generated_sentences.append(' '.join(sentence))

for sentence in generated_sentences:
    print(sentence)
```
